```python
# Fill missing flow values (if any)
horm['flow_volume'] = horm['flow_volume'].fillna('Not at all')

# Create menstruation flag
horm['is_menstruating'] = ~horm['flow_volume'].isin(['Not at all'])

# Check counts
horm['is_menstruating'].value_counts()
```

```python
horm = horm.sort_values(['id', 'day_in_study']).reset_index(drop=True)
```

```python
# Previous day's menstruation
horm['prev_is_menstruating'] = horm.groupby('id')['is_menstruating'].shift(1).fillna(False)

# Cycle start flag
horm['cycle_start'] = (horm['is_menstruating'] & (~horm['prev_is_menstruating'])).astype(int)

horm[['id','day_in_study','flow_volume','is_menstruating','prev_is_menstruating','cycle_start']].head(20)
```

```python
horm['cycle_number'] = horm.groupby('id')['cycle_start'].cumsum()

horm[['id','day_in_study','cycle_start','cycle_number']].head(20)
```

```python
horm_cycles = horm[horm['cycle_number'] > 0].copy()
```

```python
cycle_df = horm_cycles.groupby(['id', 'cycle_number']).agg({
    'day_in_study': ['min', 'max'],
    'lh': 'mean',
    'estrogen': 'mean',
    'pdg': 'mean'
}).reset_index()
```

```python
cycle_df.columns = [
    'id',
    'cycle_number',
    'cycle_start_day',
    'cycle_end_day',
    'lh_mean',
    'estrogen_mean',
    'pdg_mean'
]
```

```python
cycle_df['cycle_length'] = cycle_df['cycle_end_day'] - cycle_df['cycle_start_day'] + 1
cycle_df.head()
```

```python
cycle_df['is_irregular'] = (
    (cycle_df['cycle_length'] < 24) |
    (cycle_df['cycle_length'] > 35)
).astype(int)

cycle_df['is_irregular'].value_counts()
```

```python
cycle_df['irregular_next_cycle'] = cycle_df.groupby('id')['is_irregular'].shift(-1)

# Remove cycles where we donâ€™t know the next cycle
cycle_df = cycle_df.dropna(subset=['irregular_next_cycle']).copy()

cycle_df['irregular_next_cycle'] = cycle_df['irregular_next_cycle'].astype(int)

cycle_df[['id','cycle_number','cycle_length','is_irregular','irregular_next_cycle']].head(20)
```

```python
subject_info = pd.read_csv('subject-info.csv', low_memory=False)

cycle_df = cycle_df.merge(subject_info, on='id', how='left')
cycle_df.head()
```

```python
cycle_df.to_csv('mcphases_cycle_level.csv', index=False)
```
